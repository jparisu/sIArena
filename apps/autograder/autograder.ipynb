# AutoGrader
Evaluate one notebook (`.ipynb`) or a zip archive of notebooks (`.zip`) against a grader YAML configuration and write results to a CSV file.

```python
# ── Configuration ────────────────────────────────────────────────────────────
INPUT          = "path/to/submission.ipynb"   # .ipynb or .zip
CONFIG         = "path/to/grader.yaml"         # grader YAML config
OUTPUT         = "results.csv"                 # output CSV path
ITERATIONS     = 1                             # measurement iterations per terrain
DEBUG          = False                         # enable debug output
AUTHOR_PATTERN = r"^(?P<author>[^_]+)"         # regex to derive author from file stem
# ─────────────────────────────────────────────────────────────────────────────
```

```python
# INSTALL LIBRARY
import subprocess
import sys

SIARENA_BRANCH = "main"

# Check if installed
if "sIArena" in sys.modules:
    print("sIArena already installed.")
else:

    try:
        subprocess.check_call(["pip", "install", f"git+https://github.com/jparisu/sIArena.git@{SIARENA_BRANCH}"])
        print("sIArena Package installed.")

    except Exception as e:

        print(f"Error installing the package: {e}. Installing by magic command.")

        %pip install --upgrade git+https://github.com/jparisu/sIArena.git@main
```

```python
from sIArena.grading import grade_input_to_csv

batch_result = grade_input_to_csv(
    INPUT,
    CONFIG,
    OUTPUT,
    iterations=ITERATIONS,
    debug=DEBUG,
    author_pattern=AUTHOR_PATTERN,
)

print(f"Assignment: {batch_result.assignment_id}")
print(f"Input: {INPUT}")
print(f"Output CSV: {OUTPUT}")
print(f"Processed submissions: {len(batch_result.submission_grades)}")
```
